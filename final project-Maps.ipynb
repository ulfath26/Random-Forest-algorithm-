```python
import itertools
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

sns.set(style="whitegrid")

%matplotlib inline
```

```python
df = pd.read_csv(r'C:\Users\Welcome\Desktop\Vehicle Prediction\Accident.csv')
df.shape
```

```python
dd = df.Day_of_Week.value_counts()

plt.figure(figsize=(15,8))
plt.barh(np.arange(len(dd)), dd)
ticks = plt.yticks(np.arange(len(dd)), dd.index)
plt.xlabel('No_Of_Accidents')
```
